```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# RC-MN
d = np.linspace(-15,15,50)
df=100
conn=1*df/(df+d**2)
```

```python
print str(np.std(conn)*1.6/15) + ' mm'
```

```python
plt.figure()
plt.plot(d,conn)
plt.show()

print '15 corresponds to 1.6mm'
```

```python
# MN-RC
d = np.linspace(-2,2,50)
df=1
conn=1*df/(df+d**2)
```

```python
plt.figure()
plt.plot(d,conn)
plt.show()
print '2 corresponds to 0.213mm'
```

```python
# RC-MN
mu, sigma = 0, 3.5/3
s = np.random.normal(mu, sigma, 100)
```

```python
plt.figure()
count, bins, ignored = plt.hist(s, bins=30, density=True)
plt.plot(bins, 1/(sigma*np.sqrt(2*np.pi))*np.exp(-(bins-mu)**2/(2*sigma**2)))
plt.show()
```
